**Background**

In the early 1800s, Jean-Baptiste Joseph Fourier wanted to express solutions to the heat equation boundary value problem as infinite sums of sines and cosines (Walker, 167.) To that end, he created his namesake *Fourier series*. His work would not be appreciated until well over a century later. In today's world, scientists working in virtually every field from crystallography to circuit design make use of Fourier series and transforms. (Osgood, 1) A *Fourier series* takes the form: $$f(x)=a_1\sin(\dfrac{\pi x}{L})+b_1\cos(\dfrac{\pi x}{L})) + a_2\sin(\dfrac{2 \pi x}{L})+b_2\cos(\dfrac{2 \pi x}{L})) + ... + a_n\sin(\dfrac{n \pi x}{L})+b_n\cos(\dfrac{n \pi x}{L}))$$

With complex or real constants $a_n , b_n $. To see these series in action, we will use Fourier series to construct the aptly-named *sawtooth* wave.

The Fourier series for the sawtooth wave is derived analytically and given by:

$$ f(t) = \dfrac{A}{2}-\dfrac{A}{\pi} \sum_{n=1}^\infty \dfrac{\sin(n\omega_0t)}{n}$$

Where $\omega_0 = \dfrac{2\pi}{T}$ and $A$ is the waveform's amplitude.
